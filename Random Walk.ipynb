## From lecture 6 of the term 2 time series course, random walk defined as:

$Y_{t} = Y_{t-1} + \epsilon_{t}$,

where $t$ is part of the integers.

```r
n <- 10000
n_seq <- seq(n+1)
initial_value <- 0
y_t <- c(initial_value)
```

```r
deltas <- runif(n, -10, +10)
```

```r
for (i in deltas) {
    
    idx <- length(y_t)
    y_t <- c(y_t, y_t[idx]+i)
    
}
```

```r
plot(n_seq, y_t, type = 'l')
```

##And now adding a bias to the random walk...

> $p$ is the probability that the time series will increase, $1-p$ is the probability that the time series will decrease

```r
# probability of success
p <- 0.51

# number of steps
n <- 1000

#you're y_t timeseries process, initialised at zero
y_t <- c(0)

#sequence of t's
t <- seq(n)

#looping through the indices, adding 1 if random number is below p, and subtracting 1 if above p
for (i in t[1:n-1]) {
    
    idx <- length(y_t)
    
    if (runif(1) < p) {
        y_t <- c(y_t, y_t[idx]+1)
    }
    
    else {
        y_t <- c(y_t, y_t[idx]-1)
    }
    
}

plot(t, y_t, type='l')
```
