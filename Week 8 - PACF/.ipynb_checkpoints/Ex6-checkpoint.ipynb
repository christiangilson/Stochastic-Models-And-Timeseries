## ARGH! There are 400 observations, not 20!

* They're  just showing  the first 20
* Note that if you have 400 observations, you're going to get 399 differenced observations, as each differenced observation requires 2 observations.

```r
T.w <- 400
```

```r
se <- 1 / sqrt(T.w)

num.se <- qnorm(0.025, lower.tail = FALSE)

CI.95 <- 0 + c(-num.se*se, num.se*se)

CI.95
```

### ACF of  $\{y_{t}\}$

* Linear rather than geometric decrease suggests that ${y_{t}}$ is a non-stationary process

```r
plot(lag, y.acf)
```

## ACF of $\{W_{t}\} = \{\Delta Y_{t}\}$

* This is now a stationary process as we no longer see a linear trend
* We now see scattered points  around the zero mean
* Plotting the 95% confidence interval for a white noise process, we can see that all points lie within that confidence interval (we would expect one of the twenty points to lie outside it, and we see none).
* Therefore we identify $\{\Delta Y_{t}\}$ as a white noise process
* $\{Y_{t}\}$ is a non-stationary ARIMA(1,1,0) model:

> $W_{t} = \epsilon_{t}$

> $(1-L)Y_{t} = \epsilon_{t}$

> $Y_{t} = \phi Y_{t-1} + \epsilon_{t}$

where $\phi = 1$, and is this **non-stationary**.

```r
plot(lag, w.acf, ylim = c(-1,1))
abline(h=0)
abline(h=CI.95[1])
abline(h=CI.95[2])
```

# **INCORRECT!**

> Should have found that  there were two points and a sharp cut off, highlighting an MA(2) process.

> Then fit the confidence intervals around a suspected MA(2):

**95% confidence interval around a suspected MA(q) process**:

> $\pm 2 \sqrt{\dfrac{1 + 2(r_{1}^{2} + r_{2}^{2} + ... + r_{q}^{2})}{T}}$

**We suspect MA(2), therefore let's fit a CI around**:

> $\pm 2 \sqrt{\dfrac{1 + 2(r_{1}^{2} + r_{2}^{2})}{T}}$


```r
T <- 399
```

```r
num.se
```

```r
se <- sqrt( (1 + 2*(0.145**2 + 0.362**2)) / T)
se
```

```r
CI.95 <- 0 + c(-num.se*se, num.se*se)
```

```r
CI.95
```

```r
plot(lag, w.acf, ylim = c(-1,1))
abline(h=0)
abline(h=CI.95[1])
abline(h=CI.95[2])
```

## New Summary:

* Can see that the data  supports our MA(2) hypothesis for $W_{t}$

* Thus there's evidence to suggest $Y_{t}$ is an ARIMA(0,1,2) process.

## Q2) Sheep Data

```r
sheep <- read.csv('sheepdata.txt', header = FALSE)
colnames(sheep) <- c('Pop')

sheep.ts <- ts(sheep, start = 1867, frequency = 1)

sheep.log.ts <- log(sheep.ts)

sheep.lag.1.ts <- diff(sheep.ts, lag=1)
```

```r
plot(sheep.ts)
```

```r
plot(sheep.log.ts)
```

```r
plot(sheep.lag.1.ts)
```

## 1) Is it stationary or not?

* Have looked at sample ACFs for both the process and the log of the process, and it looks like there's a linear decrease in the ACF that suggests the process is non-stationary

```r
acf(sheep.ts, lag = 30)
```

```r
acf(sheep.log.ts, lag=30)
```

##Â Not a stationary process.

## Must apply differencing to transform into a stationary process.

```r
acf(sheep.lag.1.ts, lag=100)
```

## Strongly suggests an ARIMA model
> ACF shows geometric progression, so unlikely to be MA

> Shows values above the 95% confidence interval for a white noise process

> So the differenced process is either AR(p) or ARMA(p,q)

> Suspect it's AR(3)

```r
pacf(sheep.lag.1.ts, lag=50)
```

# Fitting ARIMA models

```r
sheep.3.1.0 <- arima(sheep.ts, order=c(3,1,0))
sheep.3.1.0
```

```r
sheep.3.1.1 <- arima(sheep.ts, order=c(3,1,1))
sheep.3.1.1
```

#Â Summary:

* Will look at the significance of the 4th parameter in ARIMA(4,1,0) to see if significantly differs from 0 in an overfitting test.

## Looking at ARIMA(4,1,0)

```r
sheep.4.1.0
```

```r
t <- -0.0481 / 0.1323
```

```r
t
```

```r
T <- length(sheep.ts)
T
```

```r
# degrees of freedom:
## T - 4 (parameters)
## - 1 (differencing so really our modified T = T-1)
## - 1 (mean) - 1 (extra)
df <-  T - 4 - 1 - 1 - 1
df
```

```r
2*pt(t, df, lower.tail=TRUE)
```

**Cannot reject  the  null hypothesis that the fourth AR parameter is equal to zero, thus we accept the alternative that $\phi_{4} = 0$**

##Â Model equation and fitted equation:

#### ARIMA(3,1,0)

> $W_{t} = (1-L)Y_{t}$

#### ARMA(3,0) for $\{W_{t}\}$

> $\phi(L)(W_{t} - \mu) = \theta(L)\epsilon_{t}$

#### For ARMA(3,0):

> $\theta(L) = 1$

> $\phi(L) = 1 - \phi_{1}L - \phi_{2}L^{2} - \phi_{3}L^{3}$

#### Full ARMA(3,0):

> $(1 - \phi_{1}L - \phi_{2}L^{2} - \phi_{3}L^{3})(1-L)Y_{t} = \epsilon_{t}$

> $Y_{t} = (1+\phi_{1})Y_{t-1} - (\phi_{1} - \phi_{2})Y_{t-2} - (\phi_{2} - \phi_{3})Y_{t-3} - \phi_{3}Y_{t-4} + \epsilon_{t}$
