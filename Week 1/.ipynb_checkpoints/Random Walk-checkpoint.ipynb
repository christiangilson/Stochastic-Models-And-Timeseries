```r
library(repr)

options(repr.plot.width=22, repr.plot.height=10)
```

```r
n = 50

p = 0.55
```

```r
s = 0
```

```r
# get values from uniform distribution

# runif gets a random sample, with a default between 0 and 1
u.values = runif(0:n)
```

## Lecturer's method

```r
# jumps
## this is kind of like a list comprehension in python
## the u.values inqualities provide TRUE or FALSE outputs in a vectorised fashion
## pick +1 for cases where it's TRUE
## pick -1 for else
jumps = 1*(u.values <= p) - 1*(u.values > p)
```

```r
# looping through range from 1 to 50, including 50
for (i in 1:n) {
    # updating s
    ## s starts at 0
    ## appending s from previous value and adding the jump which is either 1 or -1
    s = c(s, s[length(s)] + jumps[i])
}
    
```

```r
plot (s)
```

## Christian's method

```r
n = 10000

p = 0.51

s = c(0)
s_i = 0

for (i in 1:n) {
    
    # each time you loop through you generate a new random uniform number
    u = runif(1)
    
    if (u <= p) {
        delta = 1
    } else if (u > p) {
        delta = -1
    }
    
    s_i = s_i + delta
        
    s = c(s, s_i)
}

plot(s)
```
